```python
quasi_dist_x = []
quasi_dist_y = []
peaks_dist_x = []
peaks_dist_y = []
```

```python
for i, pth in enumerate((base_accumulate / "quasi" / "dis").iterdir()):
    current = np.load(pth)
    quasi_dist_x.append(current.copy())
for i, pth in enumerate((base_accumulate / "quasi" / "abs_dis").iterdir()):
    current = np.load(pth)
    quasi_dist_y.append(current.copy())
for i, pth in enumerate((base_accumulate / "peaks" / "dis").iterdir()):
    current = np.load(pth)
    peaks_dist_x.append(current.copy())
for i, pth in enumerate((base_accumulate / "peaks" / "abs_dis").iterdir()):
    current = np.load(pth)
    peaks_dist_y.append(current.copy())
```

```python
quasi_dist_x = np.concatenate(quasi_dist_x)
quasi_dist_y = np.concatenate(quasi_dist_y)
peaks_dist_x = np.concatenate(peaks_dist_x)
peaks_dist_y = np.concatenate(peaks_dist_y)
```

```python
plot_dens_scatter(quasi_dist_x, quasi_dist_y, "Dis Distribution", "dis", "abs_dis")
```

```python
plot_dens_scatter(peaks_dist_x.flatten(), peaks_dist_y.flatten(), "Peaks Dis Distribution", "dis", "abs_dis")
```

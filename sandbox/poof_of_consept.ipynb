```python
import numpy as np
import matplotlib.pyplot as plt

# Number of data points per cluster
num_points = 100

# Generate centroids
centroid = [[4, -1], [-4, 1], [-2, -5], [2, 5]]
norms = np.linalg.norm(centroid, axis=1, keepdims=True)
centroids = centroid/norms


# Generate data points around each centroid with noise
data = []

# Add noise level to the angle (adjust as needed)
angle_noise = np.pi / 18  # Adjust this value to control the amount of noise (e.g., Ï€/18 radians is approximately 10 degrees)

for centroid in centroids:
    for _ in range(num_points):
        # Generate random noise for the angle
        noise_angle = np.random.uniform(-angle_noise, angle_noise)

        # Apply noise to the angle of the centroid coordinates
        new_angle = np.arctan2(centroid[1], centroid[0]) + noise_angle
        noise_point = np.array([np.cos(new_angle), np.sin(new_angle)])

        data.append(noise_point)

# Convert the data to a NumPy array
data = np.array(data)

# Visualize or use the generated data as needed
```

```python
# Split data into clusters
cluster_1 = data[:num_points]
cluster_2 = data[num_points:2 * num_points]
cluster_3 = data[2 * num_points:3 * num_points]
cluster_4 = data[3 * num_points:4 * num_points]


# Create a scatter plot for each cluster with different colors
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], label="Cluster 1", c="r", marker="o")
plt.scatter(cluster_2[:, 0], cluster_2[:, 1], label="Cluster 2", c="g", marker="s")
plt.scatter(cluster_3[:, 0], cluster_3[:, 1], label="Cluster 3", c="b", marker="^")
plt.scatter(cluster_4[:, 0], cluster_4[:, 1], label="Cluster 4", c="c", marker="x")


# Add labels and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Clustered Data")
plt.legend()

# Display the plot
plt.show()
```

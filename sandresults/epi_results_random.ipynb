MOST BRAIN COMPONENTS, EYES CLOSED

```python
stab_folder_1 = pathlib.Path(r"C:\Users\workbench\eirik_master\Results\epi_data\accumulated_clean\1")
```

```python
quasi_dist_x = []
quasi_dist_y = []
peaks_dist_x = []
peaks_dist_y = []
```

```python
for i, pth in enumerate((stab_folder_1 / "quasi" / "dis").iterdir()):
    current = np.load(pth)
    quasi_dist_x.append(current.copy())
for i, pth in enumerate((stab_folder_1 / "quasi" / "abs_dis").iterdir()):
    current = np.load(pth)
    quasi_dist_y.append(current.copy())
for i, pth in enumerate((stab_folder_1 / "peaks" / "dis").iterdir()):
    current = np.load(pth)
    peaks_dist_x.append(current.copy())
for i, pth in enumerate((stab_folder_1 / "peaks" / "abs_dis").iterdir()):
    current = np.load(pth)
    peaks_dist_y.append(current.copy())
```

```python
quasi_dist_x = np.concatenate(quasi_dist_x)
quasi_dist_y = np.concatenate(quasi_dist_y)
peaks_dist_x = np.concatenate(peaks_dist_x)
peaks_dist_y = np.concatenate(peaks_dist_y)
```

```python
print(quasi_dist_x.shape)
print(quasi_dist_y.shape)
print(peaks_dist_x.shape)
print(peaks_dist_y.shape)
```

```python
plot_dens_scatter(quasi_dist_x, quasi_dist_y, "Dis Distribution", "dis", "abs_dis")
```

```python
plt.hist(quasi_dist_x)
plt.show()
```

```python
plt.hist(quasi_dist_y)
plt.show()
```

```python
plt.hist(peaks_dist_y)
plt.show()
```

```python
plt.hist(peaks_dist_x)
plt.show()
```

```python
plot_dens_scatter(peaks_dist_x.flatten(), peaks_dist_y.flatten(), "Peaks Dis Distribution", "dis", "abs_dis")
```

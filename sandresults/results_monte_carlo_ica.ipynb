```python
import pandas as pd
import seaborn as sns
```

```python
v1 = np.concatenate((base_theta.flatten(), my_heavy_theta.flatten()))
v2 = np.concatenate((heavy_auto_theta, prep_theta))
df = pd.DataFrame()
df["n Randomly Sampled Epochs"] = np.tile([1, 5, 9, 14, 19, 24, 29], 4)
df["Alorithm"] = np.repeat(["Untreated", "SDX", "AutoReject", "PREP"], 7)
df["Probability of Significant Difference"] = np.concatenate((v1, v2))
```

```python
sns.set_theme(style="white")
style = sns.axes_style()
style['axes.spines.top'] = False
style['axes.spines.right'] = False
style['ticks'] = True
style['xtick.bottom'] = True
style['ytick.left'] = True
sns.set_theme(style=style, font="Times New Roman")

ax = sns.lineplot(y="Probability of Significant Difference", x="n Randomly Sampled Epochs", hue="Alorithm", data=df, alpha=0.8, palette='mako', style="Alorithm", linewidth=3)
ax.set_ylabel("Probability of Significant Difference", fontdict={'size' : 18, "font":"Times New Roman"})
ax.set_xlabel("n Randomly Sampled Epochs", fontdict={'size' : 18, "font":"Times New Roman"})
ax.set_title("Theta", fontdict={'weight' : 'bold', 'size' : 22, "font":"Times New Roman"})
ax.set_yticklabels(np.round(ax.get_yticks(), 2), size = 15)
ax.set_xticklabels((np.ceil(ax.get_xticks())).astype(int), size = 15)

sns.move_legend(ax, "upper left", bbox_to_anchor=(0.7, 0.62))
plt.legend([],[], frameon=False)
for patch in ax.patches:
    print(patch)
    r, g, b, a = patch.get_facecolor()
    patch.set_facecolor((r, g, b, .3))

plt.savefig(r"C:\Users\workbench\eirik_master\code\Masteroppgave2\images\theta_art")
plt.show()
```
